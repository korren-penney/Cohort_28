```python
# The purpose of this program is to use the random module in python 
# to roll a dice for our icebreaker today. 
# Korren is going to attempt to do this live, be prepared for typos.

# import our random library
import random

# declare variables to determine our smallest and our largest number
min = 1
max = 6

# decare var to determine if we want to roll again
# set it to yes
roll_again = "yes"

# create a loop to start asking our questions
while roll_again == 'yes':
    # print a message to indicate we are rolling the die
    print("Rolling the die...")
    
    # declare var for die value and call the random number
    die_value = (random.randint(min,max))
    
    # print the value of the random number 
    print("The value is: ", die_value)
    
    # create the if statement to tell the program what to do for each number
    if die_value == 1:
        print("If you were invisible for a day what would you do?")
    elif die_value == 2:
        print("If you could choose one super power, what would it be?")
    elif die_value == 3:
        print("What is your favourite food?")
    elif die_value == 4:
        print("Who is your favourite book or movie character?")
    elif die_value == 5:
        print("What is your favourite game to play?")
    else:
        print("What is one thing you are really good at?")
        
    # Prompt the user to see if the want to roll again
    roll_again = input("Roll again? yes/no?")
    
    # Ending the program and error handling 
    if roll_again == "no":
        print("you have selected no, program terminating")
    elif roll_again != "yes" or roll_again != "no":
        print("invalid entry, program terminating")
    
# Print a message to let user know that program has ended
print("This message will only display at the end. Program terminated.")
```
